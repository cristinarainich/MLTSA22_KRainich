```python
import numpy as np
import pandas as pd 
import pylab as pl
```

```python
ts1 = np.random.randn(100)
ts2 = np.random.randn(100) + 10
ts = np.concatenate([ts1, ts2])
pl.plot(ts);
```

```python
mu = np.zeros(ts.shape[0])
mu2 = np.zeros(ts.shape[0])
for i in range(0, ts.shape[0]-1):
  mu[i] = ts[0:i+1].mean()
  mu2[i] = ts[i:].mean()

pl.plot(ts)
pl.figure()
pl.plot(mu);
pl.figure()
pl.plot(mu2);
```

```python
rat = mu-mu2
print(np.nanmin(rat))
print(np.nanmax(rat))

pl.plot(rat);

print(np.where(rat==np.nanmin(rat)))
```
